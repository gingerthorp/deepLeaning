## 코랩 사용 단축키 정리
1. enter : 수정
2. esc :  셀 저장
3. Ctrl(CMD) + M D : 셀 삭제
4. Ctrl(CMD) + M Z : 되돌리기
5. Ctrl(CMD) + Enter : 셀 실행 후 **이동하지 않음**
6. Shift + Enter : 셀 실행 후 **다음 셀로 이동**
7. Alt + Enter : 셀 실행 후 **새로운 셀 삽입**
8. Ctrl(CMD) + M M : 코드를 테스트로 변경
9. Ctrl(CMD) + M B : 코드 셀 아래 삽입

```
diabetes = load_diabetes()
```

```
print(diabetes.data.shape, diabetes.target.shape)
```

```
diabetes.data[0:3]
```

```
diabetes.target[:3]
```

```
x = diabetes.data[:,2]
y = diabetes.target
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
w = 1.0
b = 1.0
```

```
y_hat = x[0] * w + b
print(y_hat)
```

```
print(y[0])
```

```
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
print(y_hat_inc)
```

```
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```

```
w_new = w + w_rate
print(w_new)
```

```
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)

b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```
b_new = b + 1
print(b_new)
```

```
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

```
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
```

```
print(w_new, b_new)
```

```
w=1.0, b=1.0
```

```
x = diabetes.data[:,2]
y = diabetes.target

w=1.0
b=1.0

for i in range(1,100):
  before_w = w
  before_b = b
  for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b     # y_hat = x_i * 가중치 + 절편
    err = y_i - y_hat       # 오차 역전파로 가중치와 절편을 더 적절하게 업데이트.
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err
  if (before_w == w and before_b == b):
    print("break i = ", i)
    break
print(w, b)
```

```
plt.scatter(x,y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
x_new = 0.18
y_pred = x_new * w + b
print(y_pred)
```

```
plt.scatter(x,y)
plt.scatter(x_new, y_pred)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('shape value')
plt.ylabel('target value')
plt.show()
```
